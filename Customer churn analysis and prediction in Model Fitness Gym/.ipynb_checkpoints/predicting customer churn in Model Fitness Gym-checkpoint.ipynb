## Background

Model Fitness is a gym chain developing a customer interaction strategy based on analytical data.
One of the most common problems gyms and other services face is customer churn. In order to fight churn, Model Fitness has digitized a number of its customer profiles.

For a gym, it makes sense to say a customer has left if they don't come for a month. Of course, they may be travelling and will resume their visits when they return, but that's different from a typical case. Usually, if a customer joins, comes a few times, and disappears, they're unlikely to return.

### Objective
In this project, I will assist Model Fitness in analyzing their customer churn based on the data they have and coming up with a customer retention strategy.

Here's the following analysis that will be conducted:
1. Learning to predict the probability of churn (for the upcoming month) for each customer

Model Fitness provided CSV files containing data on churn for a given month and information on the month preceding it.

The dataset includes the following fields:
- `Churn` ‚Äî the fact of churn for the month in question. `0` for not churn, and `1` for churn
- `gender` ‚Äî `0` for female, and `1` for male
- `Near_Location` ‚Äî whether the user lives or works in the neighborhood where the gym is located

- `Promo_friends` ‚Äî whether the user originally signed up through a "bring a friend" offer (they used a friend's promo code when paying for their first membership)
- `Phone` ‚Äî whether the user provided their phone number
- `Age`
- `Lifetime` ‚Äî the time (in months) since the customer first came to the gym
- `Contract_period` ‚Äî 1 month, 3 months, 6 months, or 1 year
- `Month_to_end_contract` ‚Äî the months remaining until the contract expires
- `Group_visits` ‚Äî whether the user takes part in group sessions
- `Avg_class_frequency_total` ‚Äî average frequency of visits per week over the customer's lifetime
- `Avg_class_frequency_current_month` ‚Äî average frequency of visits per week over the current month
- `Avg_additional_charges_total` ‚Äî the total amount of money spent on other gym services: cafe, athletic goods, cosmetics, massages, etc.

## User Churn Prediction : Machine Learning Application

In order to fulfill the customer churn prediction task, we need to build a model that can generalize the feature to predict user churn. Machine learning can assist us in satisfying our goals. We will classify and cluster the customer to know the traits most likely exist in the churned customer.

### Metric

```python
# create metric_evaluation fuction
def metric_evaluation (model, y_true, y_pred):
    print(f'Accuracy Score: {accuracy_score(y_true, y_pred):.2f}')
    print(f'Precision Score: {precision_score(y_true, y_pred):.2f}')
    print(f'Recall Score: {recall_score(y_true, y_pred):.2f}')
```

### Preprocessing Dataset

In this stage, we will conduct data preprocessing for the input model on classification and clustering. First, we will define the features and target variables from the dataset. After that, the numerical features will be standardized with `StandardScaler().` 

The dataset will be split into train and test sets and then standardized for the classification process. Whereas, for the clustering, we will use the whole dataset to be standardized. 

#### Classification Dataset 

```python
# Define features and target variable
## drop features with multiple collinearity

X = df.drop(columns=['churn', 'remaining_period', 'avg_visit_weekly'])
y = df['churn']

## check features
X.columns
```

```python
# define numerical features
numeric_features = ['age', 'avg_add_charge', 'contract_period',
                   'lifetime', 'avg_visit_weekly_this_month']
```

```python
# split dataset into the train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# standardize the numerical features
## define model
scaler = StandardScaler()

## train and transform only the numerical features
## OPTION A
X_train_st = X_train.copy()
X_train_st[numeric_features] = scaler.fit_transform(X_train_st[numeric_features])

X_test_st = X_test.copy()
X_test_st[numeric_features] = scaler.transform(X_test_st[numeric_features])

#----------------------------------------------------------------------#
#----------------------------------------------------------------------#
## OPTION B
# X_train_st = scaler.fit_transform(X_train)
# X_test_st = scaler.transform(X_test)
```

```python
# check the result
X_train_st.head(3)
```

#### Clustering Dataset

```python
# define features
features = df.drop(['churn'], axis=1)

# define model
sc = StandardScaler()

# train model
features_st = sc.fit_transform(features)
```

### Predict User Churn : Binary Classification

In this section, we will predict the user churn by building binary classification using the logistic regression model and the random forest regressor model. The training dataset will be trained using these methods. Lastly, we will evaluate the results from both models with accuracy, precision, and recall metrics. The evaluation lets us know which model works better. 

```python
# assign and train the LogisticRegression model
lr_model = LogisticRegression(random_state=0)
lr_model.fit(X_train_st, y_train)

# predict target variable
y_pred_lr = lr_model.predict(X_test_st)
```

```python
# assign and train the RandomForestRegressor model
rf_model = RandomForestRegressor(n_estimators=100, random_state=0)
rf_model.fit(X_train_st, y_train)

# predict target variable
y_pred_rf = rf_model.predict(X_test_st)

# set threshold to convert predicted values to binary
threshold = 0.5

# convert predicted values to binary using threshold
y_pred_rf_binary = [1 if x > threshold else 0 for x in y_pred_rf]
```

```python
# evaluate the model
print('Logistic Regression model evaluation')
metric_evaluation(lr_model, y_test, y_pred_lr)
```

```python
# evaluate the model
print('Random Forest Regressor model evaluation')
metric_evaluation(rf_model, y_test, y_pred_rf_binary)
```

**FINDINGS:** <br>
- Metric evaluation on both `Logistic Regression` and `Random Forest Regressor` gave almost similar results
- However, one way to compare the performance of these models is by looking at their recall scores. The recall score measures the proportion of true positives correctly identified by the model out of all actual positive samples.
- In this case, the `Logistic Regression` model has a higher recall score of 0.82 compared to the random forest regressor model, which has a recall score of 0.80. This means that the logistic regression model may better identify positive cases than the random forest regressor model.

**RECOMMENDATIONS:**<br>
- It's better to consider other factors, such as model complexity, training time, and interpretability, to decide which model is better.
- We can still optimize the model by trying feature selection, regularization, hyperparameter tuning, ensemble methods, or re-evaluate our model using Cross-validation.

**INTERMEDIATE CONCLUSION:**<br>
- We can use the `Logistic Regression` model to predict user churn. Pass the new dataset record this month, and we can start to predict the customers who most likely churn.

### User Clustering

#### Dendrogram

```python
# pass the standardized data to the clustering model classes, linkage
linked = linkage(features_st, method='ward')
```

```python
# visualize the clustering class
plt.figure(figsize=(15,10))
dendrogram(linked, orientation='top')
plt.title('Hierarchical Clustering for GYM Customers')
plt.show()           
```

**FINDINGS**:<br>
- From the dendrogram, the number of difference color corresponds to reccomendation of cluster numbers. The suggested optimal number of clusters is four (4)

#### KMeans Model

```python
# define the k_means model with 5 clusters
km = KMeans(n_clusters=5)

# predict the clusters for observations
# we used non-standardized features to be passed on the model
labels = km.fit_predict(X)

# store cluster labels in the field of our dataset
df['cluster_km'] = labels
```

```python
# calculate the silhouette score for evaluation
print('Silhouette_score: {:.2f}'.format(silhouette_score(X, labels)))
```

**NOTE:**<br>
A silhouette score of 0.51 suggests some degree of separation between the clusters, but there may be some overlap or ambiguity in how the data points are assigned to clusters. 

However, in our context (customer churn segmentation analysis), a moderate silhouette score may be acceptable if the goal is to identify distinct customer segments based on their behaviour characteristics. In this case, even if there is some degree of overlap or ambiguity between the clusters, the identification of distinct clusters may still provide valuable insights for the marketing and business strategy.

```python
# check the cluster amount and absolut values
df.groupby('cluster_km').agg(['count','sum'])
```

#### Cluster Visualization

```python
# check cluster proportion on churn users
for i in cat_features:
    (pd
     .crosstab(index=df[i],
               columns=df['cluster_km'],
               values=df['churn'],
               aggfunc='sum',
               normalize='index'
              )
     .plot(kind='bar', stacked='True', figsize=(5,3))
    )
    plt.title(i)
    plt.show()
```

```python
# plot distribution on numerical features
for i in non_cat_features:
    sns.displot(data=df, x=i, hue='cluster_km', palette='tab10',
                alpha=0.5, kde=True, stat='density',height=3, aspect=1.5)
    plt.title(i)
    plt.show()
```

**FINDINGS:**<br>
- Based on all categorical features, more than 70% of churn users come from customer clusters `0`,`3`, and `1` since this cluster dominates the total customers.
- The cluster difference can be seen through the feature `avg_add_charge`.
- Most churn users tend to spend extra charges below around 250 USD.

#### Churn Rate

This section will calculate the churn ratio of each cluster, so that we know which cluster tends to churn

```python
# transform table before for calculating churn ratio
churn_clusters = (df
             .groupby('cluster_km')
             .agg({'churn':['sum','count']})
             .reset_index()
              )
# drop column's level and rename columns
churn_clusters.columns = churn_clusters.columns.droplevel(0)
churn_clusters.columns = ['cluster_km', 'n_churns', 'n_users']

# calculate churn ratio
churn_clusters['churn_ratio'] = churn_clusters['n_churns']/churn_clusters['n_users']

# sort descending
churn_clusters.sort_values('churn_ratio', ascending=False)
```

```python
# check the mean feature values per cluster
df.groupby('cluster_km').mean()
```

**FINDINGS:**<br>
- Customers in clusters `0` and `3` are more likely to churn compared to others (churn ratio 35% and 33%), followed by cluster `1` (22.7%) and cluster `4` (17.8%)
- Customers in cluster `2` tend to be loyal (churn ratio below 5%)

**INSIGHTS**:<br>
- Here are the general characteristics of customers who are more likely to churn: most of them have a `lifetime` below three months,  visit less than twice a week, get lower `promo_friends`, the average additional extra charge below 106 USD, and they still have remaining contract period for about four months.

## Recommendations

Here are some strategies or solutions that may help reduce customer churn for the gym:

- **Improve the onboarding process**: It's important to make sure that new customers have a positive experience when they join the gym, especially in the first 3 months. By making customers feel welcome and supported from the beginning, they may be more likely to stick with the gym in the long run.

- **Offer incentives for frequent visits**: Since customers who visit the gym less than twice a week are more likely to churn, it may be helpful to offer incentives for more frequent visits. This could include rewards for hitting certain milestones (e.g., working out 5 times a week for a month), or discounts on membership fees for those who consistently visit the gym a certain number of times per week.

- **Offer discounts or promotions***: Since customers with lower `promo_friends` are more likely to churn, offering discounts or promotions may be a way to entice them to stay. This could involve discounts on membership fees, free personal training sessions, or other perks that make the gym more appealing. 

- **Provide personalized experiences, so they are triggered to spend more (`extra charge`)**: It may be helpful to provide personalized experiences for customers based on their fitness goals, preferences, and other factors. This could include customized workout plans, nutritional guidance, or even just tailored recommendations for classes or equipment. By providing a more personalized experience, customers may feel more invested in their fitness journey and more likely to stay with the gym in the long run.
